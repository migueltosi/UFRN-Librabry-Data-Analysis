```python
def map_cdu_classification(value):
    if 0 <= value < 100:
        return 'Generalidades. Ciência e conhecimento'
    elif 100 <= value < 200:
        return 'Filosofia e psicologia'
    elif 200 <= value < 300:
        return 'Religião'
    elif 300 <= value < 400:
        return 'Ciências sociais'
    elif 400 <= value < 500:
        return 'Classe vaga. Provisoriamente não ocupada'
    elif 500 <= value < 600:
        return 'Matemática e ciências naturais'
    elif 600 <= value < 700:
        return 'Ciências aplicadas'
    elif 700 <= value < 800:
        return 'Belas artes'
    elif 800 <= value < 900:
        return 'Linguagem. Língua. Linguística'
    elif 900 <= value < 1000:
        return 'Geografia. Biografia. História'
    else:
        return ''

df['CDU'] = df[df['localizacao'].notnull()]['localizacao'].astype(int).map(map_cdu_classification)
```

```python
categorical_df = df[['tipo_vinculo_usuario', 'biblioteca', 'colecao', 'CDU']]
def create_categorical_df(categorical_value):
  categorical_value = str(categorical_value)
  by_categorical = (categorical_df[f'{categorical_value}'].value_counts()).reset_index()

  by_categorical.columns = [f'{categorical_value}', 'quantidade']
  by_categorical['percentual'] = (((by_categorical['quantidade'] / by_categorical['quantidade'].sum())*100).round(2)).astype('string') + '%'
  return by_categorical
```

```python
by_user_link_type = create_categorical_df('tipo_vinculo_usuario')
by_library = create_categorical_df('biblioteca')
by_colecao = create_categorical_df('colecao')
by_cdu = create_categorical_df('CDU')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def create_boxplot(data,x,y,title=''):

  sns.set_theme(style="darkgrid", palette='Blues',font_scale=1.3)
  plt.figure(figsize=(16,10))
  ax = sns.boxplot(y= y, x= x, data= data)
  ax.set_title(title)
  ax = ax
```

```python
alunos_graduacao = df.query('tipo_vinculo_usuario == "ALUNO DE GRADUAÇÃO"')
alunos_graduacao.colecao.value_counts()
```

```python
alunos_graduacao_acervo_circulante = pd.DataFrame(alunos_graduacao.query('colecao == "Acervo Circulante"'))
alunos_graduacao_acervo_circulante['data_emprestimo'] = pd.to_datetime(alunos_graduacao_acervo_circulante['data_emprestimo'])
alunos_graduacao_acervo_circulante['ano'] = alunos_graduacao_acervo_circulante['data_emprestimo'].dt.year
alunos_graduacao_acervo_circulante['mes'] = alunos_graduacao_acervo_circulante['data_emprestimo'].dt.month
alunos_graduacao_acervo_circulante = alunos_graduacao_acervo_circulante[['ano','mes']]
alunos_graduacao_acervo_circulante = alunos_graduacao_acervo_circulante.value_counts().to_frame('quantidade').reset_index()
alunos_graduacao_acervo_circulante
```

```python
create_boxplot(alunos_graduacao_acervo_circulante, 'ano', 'quantidade', title= 'Distribuição dos empréstimos mensais realizados pelos alunos de graduação ao longo dos anos de 2010 a 2020 | Distribution of monthly loans taken out by undergraduate students over the years 2010 to 2020')
```
