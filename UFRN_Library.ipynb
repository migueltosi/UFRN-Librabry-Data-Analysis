```python
def map_cdu_classification(value):
    if 0 <= value < 100:
        return 'Generalidades. Ciência e conhecimento'
    elif 100 <= value < 200:
        return 'Filosofia e psicologia'
    elif 200 <= value < 300:
        return 'Religião'
    elif 300 <= value < 400:
        return 'Ciências sociais'
    elif 400 <= value < 500:
        return 'Classe vaga. Provisoriamente não ocupada'
    elif 500 <= value < 600:
        return 'Matemática e ciências naturais'
    elif 600 <= value < 700:
        return 'Ciências aplicadas'
    elif 700 <= value < 800:
        return 'Belas artes'
    elif 800 <= value < 900:
        return 'Linguagem. Língua. Linguística'
    elif 900 <= value < 1000:
        return 'Geografia. Biografia. História'
    else:
        return ''

df['CDU'] = df[df['localizacao'].notnull()]['localizacao'].astype(int).map(map_cdu_classification)
```

```python
df['matricula_ou_siape'] = df['matricula_ou_siape'].fillna('Não computado').astype(str)
```

```python
copies_date = pd.DataFrame(df['data_emprestimo'].value_counts()).reset_index()
copies_date.columns = ['data', 'quantidade']
copies_date['data'] = pd.to_datetime(copies_date['data'])
copies_date['ano'] = copies_date['data'].dt.year
copies_date['mes'] = copies_date['data'].dt.month
copies_date['dia'] = copies_date['data'].dt.day
copies_date['horario'] = copies_date['data'].dt.hour
copies_date
```

```python
copies_by_year = copies_date[['ano', 'quantidade']].groupby('ano').sum()
copies_by_year
```

```python
ax = sns.lineplot(data = copies_by_year, x = 'ano', y = 'quantidade')
ax.set_title('Número de exemplares emprestados em função do ano de 2010 a 2020 | Number of borrowed copies according to the year 2010 to 2020')
ax = ax
```

Nota-se uma notável queda no número dos exemplares emprestados, principalmente a partir de 2017 - 2020. O pico foi entre 2012 e 2014.

*(OBS: foram coletados somente os dados do primeiro semestre de
2020, portanto, tal ano possui uma queda menor do que a representada pelo gráfico)*

---

There is a notable drop in the number of borrowed copies, especially from 2017 - 2020. The peak was between 2012 and 2014.

*(NOTE: only data from the first half of

## Day 4 - Exploratory Data Analysis and Categorical Variables

```python
df
```

```python
categorical_df = df[['tipo_vinculo_usuario', 'biblioteca', 'colecao', 'CDU']]
def create_categorical_df(categorical_value):
  categorical_value = str(categorical_value)
  by_categorical = (categorical_df[f'{categorical_value}'].value_counts()).reset_index()

  by_categorical.columns = [f'{categorical_value}', 'quantidade']
  by_categorical['percentual'] = (((by_categorical['quantidade'] / by_categorical['quantidade'].sum())*100).round(2)).astype('string') + '%'
  return by_categorical
```

```python
by_user_link_type = create_categorical_df('tipo_vinculo_usuario')
by_library = create_categorical_df('biblioteca')
by_colecao = create_categorical_df('colecao')
by_cdu = create_categorical_df('CDU')
```

```python
by_user_link_type
```

  É nítido que os alunos de graduação são de longe os que fazem empréstimo de mais exemplares (78%), com alunos de pós-graduação muito abaixo, em segundo lugar (14%), sendo a somatória de todos os outros alunos correspondente a 8% dos exemplares emprestados, somente.

---

It is clear that undergraduate students are by far the ones who borrow the most copies (78%), with postgraduate students far below, in second place (14%), with the sum of all other students corresponding to 8 % of copies borrowed, only.
