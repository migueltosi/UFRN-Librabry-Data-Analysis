```python
df = merged_df[['id_emprestimo', 'codigo_barras', 'data_renovacao', 'data_emprestimo', 'data_devolucao', 'matricula_ou_siape', 'tipo_vinculo_usuario', 'Ano', 'id_exemplar', 'colecao', 'biblioteca', 'status_material', 'localizacao']]
```

```python
def map_cdu_classification(value):
    if 0 <= value < 100:
        return 'Generalidades. Ciência e conhecimento'
    elif 100 <= value < 200:
        return 'Filosofia e psicologia'
    elif 200 <= value < 300:
        return 'Religião'
    elif 300 <= value < 400:
        return 'Ciências sociais'
    elif 400 <= value < 500:
        return 'Classe vaga. Provisoriamente não ocupada'
    elif 500 <= value < 600:
        return 'Matemática e ciências naturais'
    elif 600 <= value < 700:
        return 'Ciências aplicadas'
    elif 700 <= value < 800:
        return 'Belas artes'
    elif 800 <= value < 900:
        return 'Linguagem. Língua. Linguística'
    elif 900 <= value < 1000:
        return 'Geografia. Biografia. História'
    else:
        return ''

df['CDU'] = df[df['localizacao'].notnull()]['localizacao'].astype(int).map(map_cdu_classification)
```

```python
df['matricula_ou_siape'] = df['matricula_ou_siape'].fillna('Não computado').astype(str)
```

## Day 3 - Exploratory Data Analysis and DateTime

A diretoria da biblioteca gostaria de entender se a quantidade de empréstimos está diminuindo, aumentando ou permanecendo igual ao decorrer dos últimos anos.

The library board would like to understand whether the amount of lending is decreasing, increasing, or staying the same over the last few years.

```python
loan_by_year = df.groupby('Ano').size()
loan_by_year
```

Há uma suspeita interna de que os meses com maior número de exemplares emprestados sejam março e setembro, mas não foi realizada uma análise real sobre isso.

Portanto, gere uma tabela com a quantidade total de exemplares emprestados por mês e descubra quais meses são os que possuem a maior quantidade de empréstimos realizados. Plote um gráfico de linhas.

---


The board would also like to better manage the library's human resources in line with existing work demand.

There is an internal suspicion that the months with the highest number of copies borrowed are March and September, but no real analysis was carried out on this.

Therefore, generate a table with the total number of copies borrowed per month and find out which months have the highest number of loans made. Plot a line graph.

```python
from datetime import datetime
```

```python
df['data_emprestimo'] = pd.to_datetime(df['data_emprestimo'])
```

```python
df['mes'] = df['data_emprestimo'].dt.month
```

```python
loan_by_month = df.groupby('mes').size()
loan_by_month
```

```python
print('Número de empréstimos em função dos meses do ano | Number of loans according to the months of the year')
loan_by_month.plot()
```

É evidente a recaída dos empréstimos no mês de Julho, Dezembro e Janeiro, principalmente; tanto quanto as altas nos meses de Março e Agosto.
Tal fator remete aos rumores, que acertaram em março como alta dos empréstimos, mas falharam por um mês ao dizer que setembro era o segundo maior mês de alta.
Sendo assim, deve-se buscar a exerção de atividades que não sejam de atendimento ao usuário e a programação de férias dos colaboradores nos meses 1, 7 e 12.


---

It is evident the relapse of loans in July, December and January, mainly; as much as the highs in the months of March and August.
This factor refers to the rumors, which were right in March as a rise in loans, but failed for a month when saying that September was the second biggest month of rise.
Therefore, one should seek to carry out activities that are not user service and schedule vacations for employees in months 1, 7 and 12.

Além do gerenciamento anual das atividades, a diretoria também necessita que seja planejada uma programação diária das atividades. Por este motivo, verifique quais foram os horários com maior quantidade de empréstimos ao longo de um dia inteiro.

In addition to the annual management of activities, the board also needs a daily schedule of activities to be planned. For this reason, check which were the times with the highest amount of loans over the course of a whole day.

```python
df['horario'] = df['data_emprestimo'].dt.hour
```

```python
loans_by_time = df.groupby('horario').size()
loans_by_time
```

```python
print('Número de empréstimos em função dos horários do dia | Number of loans according to time of day')
loans_by_time.plot()
```

Quase não há empréstimos de livros antes das 7 e depois das 21. O horário de pico é das 10 às 11 e das 16 (maior) até às 18, com pico relativo as 20. Dentre os outros, horários de baixa são das 7 às 9, das 12 às 13, 19 às 21, relativamente.
